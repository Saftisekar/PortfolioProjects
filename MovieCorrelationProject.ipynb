```python
#Import Libraried

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Reading Data
df = pd.read_excel(r'C:\Users\sa4wa\Downloads\Movies\Moviess.xlsx')
```

```python
# Checking for missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print("{} - {}%".format(col,pct_missing))
    
```

```python
#dropping missing values
df= df.dropna()
```

```python
#data types of all Columns
df.dtypes
```

```python
#Changing datatype to correct format

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#Creating a new column with the corrected year released 

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df['yearcorrect']
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', False)
```

```python
df
```

```python
df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Creating Scatter plot to see a corelation between budget and gross earnings

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget')

plt.show()
```

```python
#Plotting regression plot using seaborn

sns.regplot(x= "budget", y = "gross", data = df ,scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"})
```

```python
#Looking for correlation between budget and gross
df.corr(method = 'pearson')
```

```python
# Budget and gross have a high correlation
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap (correlation_matrix , annot = True)
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap (correlation_matrix , annot = True)

plt.title('Corelation matrix for Numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
corr_matrix = df_numerized.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
